```python
import random
import numpy as np
from lxml import etree
from datetime import datetime
```

```python
# Load the XML file
tree = etree.parse('burma14.xml')                  # Change the name between burma and Brazil and to get desired output
elem = tree.getroot()
```

```python
cities = elem.findall('graph/vertex')
# Get the number of cities
num_cities = len(cities)
```

```python
num_cities
```

```python
# Initialize a 2-D array to store the distances between cities
distance_matrix = [[0] * num_cities for _ in range(num_cities)]
```

```python
distance_matrix
```

```python
# Iterate through the XML data and fill the distance matrix
for city_id, city in enumerate(cities):
    routes = city.findall('edge')
    for route in routes:
        target_city_id = int(route.text)
        cost = float(route.get("cost"))
        distance_matrix[city_id][target_city_id] = cost
```

```python
distance_matrix
```

```python
# Print the distance matrix
for i in range(num_cities):
    for j in range(num_cities):
        print(f'Distance from city {i} to city {j} = {distance_matrix[i][j]}')
```

```python
# Parameters
n_cities = num_cities  # The number of cities from the XML data
n_population = 400
mutation_rate = 0.2
```

```python
# First step: Create the first population set of 100, randomly ordered
def genesis(n_population, n_cities):
    population_set = []
    for i in range(n_population):
        sol_i = np.random.permutation(n_cities)  # Randomly shuffle the cities
        print(sol_i)
        population_set.append(sol_i)
    return np.array(population_set)
```

```python
population_set = genesis(n_population, n_cities)      # Randomly ordered 100 city routes (aka population set of 100 routes)
```

```python
population_set
```

```python
def compute_city_distance(a, b):
  return distance_matrix[a][b]
```

```python
def fitness_eval(city_list):
    total = 0
    for i in range(n_cities - 1):
        a = city_list[i]
        b = city_list[i + 1]
        total += compute_city_distance(a, b)
    # Add distance from the last city back to the starting city
    total += compute_city_distance(city_list[-1], city_list[0])
    return total
```

```python
def get_all_fitnes(population_set):
    fitnes_list = np.zeros(n_population)

    for i in range(n_population):
        fitnes_list[i] = fitness_eval(population_set[i])

    return fitnes_list
```

```python
# Get the fitness score for each of the routes in the population set
# and store them in a list of 100 entries. [Each entry corresponds to
# the cost (aka fitness score) to take that route and be back to
# starting city.]
fitnes_list = get_all_fitnes(population_set)
```

```python
# When we print this list, each entry corresponds to
# the cost of traveling in that route. This is also
# fitness score. So lower the score, better the route.
fitnes_list
```

```python
# Now that we have score for each of the 100 populated routes, we will
# pick the best routes from the lot.

def progenitor_selection(population_set,fitnes_list):
    fitnes_list=(fitnes_list.max()-fitnes_list)/fitnes_list.sum()
    total_fit = fitnes_list.sum()
    prob_list = fitnes_list/total_fit

    # print (f'total_fit = {total_fit}\n')
    # print (f'prob_list = {prob_list}\n')

    #Notice there is the chance that a progenitor. mates with oneself
    progenitor_list_a = np.random.choice(list(range(len(population_set))), len(population_set),p=prob_list, replace=True)
    progenitor_list_b = np.random.choice(list(range(len(population_set))), len(population_set),p=prob_list,replace=True)

    # print (f'progenitor_list_a [before] = {progenitor_list_a}\n')
    # print (f'progenitor_list_b [before] = {progenitor_list_b}\n')


    progenitor_list_a = population_set[progenitor_list_a]
    progenitor_list_b = population_set[progenitor_list_b]

    # print (f'progenitor_list_a [after] = {progenitor_list_a}\n')
    # print (f'progenitor_list_b [after] = {progenitor_list_b}\n')


    return np.array([progenitor_list_a,progenitor_list_b])


progenitor_list = progenitor_selection(population_set,fitnes_list)
progenitor_list[0][2]
```

```python
# Crossover fn.
# Crossover is done as follows:
# Select first 6 cities from progenitor list1
# Select next set of cities from progenitor list2, but
# only if the city was not already picked from list1

def mate_progenitors(prog_a, prog_b):
    offspring = prog_a[0:5]

    for city in prog_b:

        if not city in offspring:
            offspring = np.concatenate((offspring,[city]))

    return offspring


# We receive progenitor_list which is a 2x100 array set.
# Among this, we select the the higher order array set
# i.e. progenitor_list[1][100]
# And then create a new population set of 100 city routes, again.

def mate_population(progenitor_list):
    new_population_set = []
    # print('Length of progenitor_list = ' + str(len(progenitor_list)))
    # print(f'progenitor_list = {progenitor_list}')
    # print(f'progenitor_list.shape[0] = {progenitor_list.shape[0]}')
    # print(f'progenitor_list.shape[1] = {progenitor_list.shape[1]}')

    for i in range(progenitor_list.shape[1]):
        prog_a, prog_b = progenitor_list[0][i], progenitor_list[1][i]
        offspring = mate_progenitors(prog_a, prog_b)
        new_population_set.append(offspring)

    return new_population_set

new_population_set = mate_population(progenitor_list)
print(new_population_set)
new_population_set[0]
```

```python
def mutate_offspring(offspring):
    for q in range(int(n_cities*mutation_rate)):
        a = np.random.randint(0,n_cities)
        b = np.random.randint(0,n_cities)

        # print(f'a = {a}, b = {b}')

        # Swap the two cities a & b
        # print(f'[Before] offspring[a] = {offspring[a]}, offspring[b] = {offspring[b]}')
        offspring[a], offspring[b] = offspring[b], offspring[a]
        # print(f'[After] offspring[a] = {offspring[a]}, offspring[b] = {offspring[b]}')

    return offspring


def mutate_population(new_population_set):
    mutated_pop = []
    for offspring in new_population_set:
        mutated_pop.append(mutate_offspring(offspring))
    return mutated_pop



# With new population set now available through cross over,
# we will now go through the mutation process.
mutated_pop = mutate_population(new_population_set)
print('-' * 60)
print(mutated_pop[0])
print('-' * 60)
```

```python
# Finally, let us look at the best solution for the
# mutated population again.

# Initialize best_solution
best_solution = [-1,np.inf,np.array([])]

best_route = []

# Loop through 10000 iterations for the genetic algo. solution to converge
for i in range(10000):
    # Print every 100th loop
    if i%100==0: print(i, fitnes_list.min(), fitnes_list.mean(), datetime.now().strftime("%d/%m/%y %H:%M"))
    fitnes_list = get_all_fitnes(mutated_pop)

    #Saving the best solution if current value is better than previously saved one
    if fitnes_list.min() < best_solution[1]:
        best_solution[0] = i
        best_solution[1] = fitnes_list.min()
        best_solution[2] = np.array(mutated_pop)[fitnes_list.min() == fitnes_list]

    # Repeat crossover for currently iterated population data
    progenitor_list = progenitor_selection(population_set,fitnes_list)
    new_population_set = mate_population(progenitor_list)

    # Repeat mutation for currently iterated population data
    mutated_pop = mutate_population(new_population_set)
```
